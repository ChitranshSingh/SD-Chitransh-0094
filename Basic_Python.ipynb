#Looping

```python
l = [1, 2, "heel", 45, 'b', '56', 38, 553, 4.5]
```

```python
num = []
strr = []

for i in l:
  if type(i) == int or type(i) == float:
    num.append(i)
  else:
    strr.append(i)

print(num)
print(strr)
```

```python
l = [12, 43, "hien", "babyyyyy", 45.4, False, 67.3, True]
```

```python
num = []
strr = []
boolean = []
floatt = []

for i in l:
  if type(i) == int:
    num.append(i)
  elif type(i) == str:
    strr.append(i)
  elif type(i) == float:
    floatt.append(i)
  elif type(i) == bool:
    boolean.append(i)

print(num)
print(strr)
print(boolean)
print(floatt)
```

##Palindrome Numbers

```python
#input a number from user
num = input("Enter a number: ")
if (num == num[::-1]):
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
x = int(input("Enter a number: "))
rev = 0;
temp = x;

while x>0:
  rev = rev*10 + x%10
  x = x//10

if(temp == rev):
  print("Palindrome")
else:
  print("Not Palindrome")
```
